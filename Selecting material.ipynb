```python
def encode(a, y, n):
    gram_2, t = ngram(a, y, n)
    str_df_1 = []
    str_df = []
    targ = []
    for i in range(len(gram_2)):
        for s in range(len(gram_2[i])):
        #print(i, s)
        #print(gram_2[i][s])
        #print(len(gram_2[i][s]))
            if gram_2[i][s].isnumeric(): is_num = 1
            else: is_num = 0
        #print(is_num)
            str_df_1.append([len(gram_2[i][s]), is_num])
            if gram_2[i] == t: target = 1
            else: target = 0
        targ.append(target)    
    str_df.append(str_df_1)
    
    dr = np.reshape(str_df, (len(gram_2), n*2))

    cols = []
    for s in range(n):
    #print(s+1)
        col_len = 'Gr_' + str(n) + '_' + str(s+1) + '_len'
        col_isnum = 'Gr_' + str(n) + '_' + str(s+1) + '_isnum'
        cols.append(col_len)
        cols.append(col_isnum)
    
    frame = pd.DataFrame(dr, columns = cols)    
    tar = pd.DataFrame(np.reshape(targ, (len(gram_2), 1)), columns = ['target']) 
    
    fr = pd.concat([frame, tar], axis = 1)
    
    frst = []
    last = []

    for i in range(dr.shape[0]):
    #for s in range(dr.shape[1]):
        if i == 0: 
            first = [0,0]
        #continue
        if i != 0:
            first = [dr[i-1, 0], dr[i-1, 1]]
        
        if i != dr.shape[0]-1:
            lst = [dr[i+1, dr.shape[1]-2], dr[i+1, dr.shape[1]-1]]
        if i == dr.shape[0]-1:
            lst = [0,0]
        frst.append(first)
        last.append(lst)

    first_df = pd.DataFrame(frst, columns = ['first_len', 'first_isnum'])
    last_df = pd.DataFrame(last, columns = ['last_len', 'last_isnum'])
    f_l = pd.concat([first_df, last_df], axis = 1)
    
    f_l['n-gram'] = n
    
    return pd.concat([f_l, fr], axis = 1)    
```

```python
def encode5(X, y):    
    a2 = encode(X, y, 2)
    a3 = encode(X, y, 3)
    a4 = encode(X, y, 4)
    a5 = encode(X, y, 5)
    return pd.concat([a2, a3, a4, a5], axis = 0).fillna(0)
```
